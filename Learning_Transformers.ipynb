```python
from transformers import pipeline
import pandas as pd
```

**Text Classification Model - Sentiment Analysis**

```python
classifier = pipeline("text-classification", model = "SamLowe/roberta-base-go_emotions")

text = '''
I walked into the café today, expecting my usual serene retreat, only to be met with the deafening screech of a malfunctioning espresso machine. The barista, bless his heart, tried to compensate for the chaos with an exaggeratedly cheerful smile, which somehow only made things worse. I ordered my usual caramel macchiato, and to my surprise, they got my name right for the first time in months—though they spelled it with a ‘Q’ instead of a ‘C.’

The first sip was... an experience. The overwhelming bitterness clashed violently with the supposed caramel sweetness, like a symphony where all the instruments were slightly out of tune. But hey, at least it was hot—scalding, in fact, because what’s a morning coffee without the added thrill of potential third-degree burns?

Despite the assault on my taste buds, I had a moment of peace. The sun filtered through the large windows, casting golden patterns on the wooden tables, and for a fleeting second, all was well. Then, of course, my bliss was shattered by the unmistakable sound of someone attempting to explain cryptocurrency to a visibly disinterested date at the next table.

By the time I left, I couldn’t decide whether I was more annoyed, amused, or genuinely impressed by the sheer absurdity of it all. But I suppose that’s just life—one bad coffee away from a philosophical crisis.
'''

outputs = classifier(text)
pd.DataFrame(outputs)
```

**Text Generation**

```python
from transformers import set_seed
set_seed(0)
```

```python
text = '''
I walked into the café today, expecting my usual serene retreat, only to be met with the deafening screech of a malfunctioning espresso machine. The barista, bless his heart, tried to compensate for the chaos with an exaggeratedly cheerful smile, which somehow only made things worse. I ordered my usual caramel macchiato, and to my surprise, they got my name right for the first time in months—though they spelled it with a ‘Q’ instead of a ‘C.’

By the time I left, I couldn’t decide whether I was more annoyed, amused, or genuinely impressed by the sheer absurdity of it all. But I suppose that’s just life—one bad coffee away from a philosophical crisis.
'''

generator = pipeline("text-generation", model="gpt2-large")
response = "I am sorry to hear about your issue."

prompt = "user: " + text + "\n Customer Service Response: " + response

outputs = generator(prompt, max_length = 256)
outputs
```

**Image Classification**

```python
from PIL import Image
import requests
from transformers import pipeline
```

```python
url = "https://images.pexels.com/photos/1108099/pexels-photo-1108099.jpeg?auto=compress&cs=tinysrgb&w=1260&h=750&dpr=2"
image = Image.open(requests.get(url, stream=True).raw)

classifier = pipeline("image-classification")
```

```python
output = classifier(image)
output
```

**IMage Segmentation**

```python
url = "https://www.shutterstock.com/image-photo/beach-seat-laptop-wallpepar-hd-260nw-2498994455.jpg"
image = Image.open(requests.get(url, stream=True).raw)

segmenter = pipeline("image-segmentation", model="mattmdjaga/segformer_b2_clothes")
output = segmenter(image)
output
```

**Text to Speech**

```python
import soundfile as sf


text = '''
I walked into the café today, expecting my usual serene retreat, only to be met with the deafening screech of a malfunctioning espresso machine. The barista, bless his heart, tried to compensate for the chaos with an exaggeratedly cheerful smile, which somehow only made things worse. I ordered my usual caramel macchiato, and to my surprise, they got my name right for the first time in months—though they spelled it with a ‘Q’ instead of a ‘C.’

By the time I left, I couldn’t decide whether I was more annoyed, amused, or genuinely impressed by the sheer absurdity of it all. But I suppose that’s just life—one bad coffee away from a philosophical crisis.
'''

synth = pipeline("text-to-speech")
```

```python
speech = synth(text)
```

```python
sf.write("speech.wav", speech["audio"].T, samplerate=['sampling_rate'])
```
